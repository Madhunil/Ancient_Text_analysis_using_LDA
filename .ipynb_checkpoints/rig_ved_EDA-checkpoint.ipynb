```python
import bs4 as bs
import urllib.request
import re
import nltk
import chars2vec

import requests
from bs4 import BeautifulSoup
```

```python
def get_links(base_url,start_index,end_index):
  urls=[]
  for i in range(start_index,end_index+1):
    temp=base_url+str(i)+".htm"
    urls.append(temp)
  return urls
```

```python
def scrape_p_1(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.content, "html.parser")
  text = soup.find_all('p')[1].prettify()
  my_list = [i.replace('<br>', '').replace('<br/>', '').replace('<p>', '').replace('</p>', '') for i in text.split()]
  return ' '.join(my_list)
```

```python
def get_text_list(urls):
  all_text=[]
  for url in urls:
    all_text.append(scrape_p_1(url))
  return all_text
```

```python
def get_text_list2(urls):
  all_text=''
  for url in urls:
    all_text = all_text + scrape_p_1(url)+"\n"
  return all_text
```

```python
(book1_pp)
```

```python
# Removing Stop Words
from nltk.corpus import stopwords
for i in range(len(book1_pp)):
    book1_pp[i] = [w for w in book1_pp[i] if w not in stopwords.words('english')]
```
